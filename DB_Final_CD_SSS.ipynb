```python
listaStats=[]
ii=0

while ii<len(listaStatsProv2017):
    numList=0
    listaStats=[]
    while numList<len(listaStatsProv2017[ii]):
        important=listaStatsProv2017[ii][numList]+listaStatsProv2018[ii][numList]
        if numList==0:
            listaStats.append(listaStatsProv2017[ii][numList])
        elif numList==1:
            listaStats.append(listaStatsProv2017[ii][numList])
        elif numList==2:
            listaStats.append(listaStatsProv2017[ii][numList])
        elif numList==5:
            if listaStatsProv2018[ii][4]+listaStatsProv2017[ii][4]==0:
                ppgF=0
            else:
                ppgF=round((listaStatsProv2017[ii][numList]*listaStatsProv2017[ii][4]+listaStatsProv2018[ii][numList]*listaStatsProv2018[ii][4])/(listaStatsProv2018[ii][4]+listaStatsProv2017[ii][4]),2)
            listaStats.append(ppgF)
        else:
            if important != str(important):
                num=round(important,2)
                listaStats.append(num)
        numList+=1
    listaStats.append(team)
    if listaStats[2]=='Goalkeeper':
        listaNulaGK.append(listaStats)
    else:
        listaNula.append(listaStats)
    ii+=1
```
